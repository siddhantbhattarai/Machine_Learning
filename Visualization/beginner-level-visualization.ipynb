```python
import seaborn as sns
iris = sns.load_dataset('iris')
```

**Line Plot**
- Plot sepal length for the first 10 entries.

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 6))
plt.plot(iris['sepal_length'][:10], marker='o')
plt.title('Sepal Length for First 10 Iris Entries')
plt.xlabel('Sample Index')
plt.ylabel('Sepal Length (cm)')
plt.grid(True)
plt.show()
```

**Scatter Plot**
- Plot sepal length vs. sepal width.

```python
plt.figure(figsize=(10, 6))
plt.scatter(iris['sepal_length'], iris['sepal_width'], c='blue')
plt.title('Sepal Length vs Sepal Width')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.grid(True)
plt.show()
```

**Bar Chart**
- Plot the count of each species.

```python
species_counts = iris['species'].value_counts()
plt.figure(figsize=(10, 6))
plt.bar(species_counts.index, species_counts.values, color=['red', 'green', 'blue'])
plt.title('Count of Each Species')
plt.xlabel('Species')
plt.ylabel('Count')
plt.show()
```

**Example 4: Pie Chart**
- Create a pie chart showing the proportion of each species in the Iris dataset.

```python
species_counts = iris['species'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(species_counts, labels=species_counts.index, autopct='%1.1f%%', startangle=140, colors=['lightblue', 'lightgreen', 'lightcoral'])
plt.title('Proportion of Each Species in the Iris Dataset')
plt.show()
```

**Example 5: Horizontal Bar Chart**
- Create a horizontal bar chart for the average sepal length of each species.

```python
avg_sepal_length = iris.groupby('species')['sepal_length'].mean()
plt.figure(figsize=(10, 6))
plt.barh(avg_sepal_length.index, avg_sepal_length.values, color=['lightblue', 'lightgreen', 'lightcoral'])
plt.title('Average Sepal Length by Species')
plt.xlabel('Average Sepal Length (cm)')
plt.ylabel('Species')
plt.show()
```

**Example 6: Histogram with Multiple Variables**
- Create histograms for sepal length and sepal width on the same plot.

```python
plt.figure(figsize=(10, 6))
plt.hist(iris['sepal_length'], bins=30, alpha=0.5, label='Sepal Length', color='blue')
plt.hist(iris['sepal_width'], bins=30, alpha=0.5, label='Sepal Width', color='green')
plt.title('Distribution of Sepal Length and Sepal Width')
plt.xlabel('Length/Width (cm)')
plt.ylabel('Frequency')
plt.legend(loc='upper right')
plt.show()
```

**Dataset**
- Titanic Dataset - Available from Kaggle or the Seaborn dataset library.

**Example 1: Simple Line Plot**
- Plot the cumulative sum of passenger ages

```python
import matplotlib.pyplot as plt
import seaborn as sns

titanic = sns.load_dataset('titanic')
age_cumsum = titanic['age'].dropna().cumsum()

plt.figure(figsize=(10, 6))
plt.plot(age_cumsum, marker='o')
plt.title('Cumulative Sum of Passenger Ages')
plt.xlabel('Passenger Index')
plt.ylabel('Cumulative Age')
plt.grid(True)
plt.show()
```

**Example 2: Basic Scatter Plot**
- Plot age vs. fare.

```python
plt.figure(figsize=(10, 6))
plt.scatter(titanic['age'], titanic['fare'], c='blue', alpha=0.5)
plt.title('Age vs Fare')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.grid(True)
plt.show()
```

**Example 3: Bar Chart**
- Plot the count of survivors and non-survivors.

```python
survival_counts = titanic['survived'].value_counts()
plt.figure(figsize=(10, 6))
plt.bar(['Non-Survivor', 'Survivor'], survival_counts, color=['red', 'green'])
plt.title('Count of Survivors and Non-Survivors')
plt.xlabel('Survival Status')
plt.ylabel('Count')
plt.show()
```

**Dataset**
- World Happiness Report Dataset - Available from Kaggle.

**Example 1: Bar Chart**
- Plot the top 10 countries by happiness score.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
happiness = pd.read_csv('https://raw.githubusercontent.com/siddhantbhattarai/Machine_Learning/main/Datasets/World-Happiness-Report.csv')

# Top 10 countries by happiness score
top10_happiness = happiness.nlargest(10, 'Happiness Score')

plt.figure(figsize=(12, 6))
plt.bar(top10_happiness['Country'], top10_happiness['Happiness Score'], color='skyblue')
plt.title('Top 10 Countries by Happiness Score')
plt.xlabel('Country')
plt.ylabel('Happiness Score')
plt.xticks(rotation=45)
plt.show()
```

- Plot the happiness score distribution


```python
import matplotlib.pyplot as plt

# Plot the happiness score distribution
plt.figure(figsize=(10, 6))
plt.hist(happiness['Happiness Score'], bins=20, color='skyblue', edgecolor='black')
plt.title('Happiness Score Distribution')
plt.xlabel('Happiness Score')
plt.ylabel('Frequency')
plt.show()
```

2. Create a bar chart of the top 10 happiest countries

```python
# Sort the dataset by Happiness Score in descending order and select the top 10 countries
top_10_happiest = happiness.sort_values(by='Happiness Score', ascending=False).head(10)

# Create a bar chart
plt.figure(figsize=(12, 6))
plt.bar(top_10_happiest['Country'], top_10_happiest['Happiness Score'], color='green')
plt.title('Top 10 Happiest Countries')
plt.xlabel('Country')
plt.ylabel('Happiness Score')
plt.xticks(rotation=45)
plt.show()
```

3. Scatter plot of Economy vs. happiness score

```python
# Create a scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(happiness['Economy'], happiness['Happiness Score'], color='purple')
plt.title('Economy vs. Happiness Score')
plt.xlabel('Economy')
plt.ylabel('Happiness Score')
plt.grid(True)
plt.show()
```
