**Dataset**
- Tips Dataset - Available from the Seaborn dataset library.

**Histogram**
- Plot the distribution of total bill amounts.

```python
import matplotlib.pyplot as plt
import seaborn as sns
tips = sns.load_dataset('tips')

plt.figure(figsize=(10, 6))
plt.hist(tips['total_bill'], bins=30, color='purple', edgecolor='black')
plt.title('Distribution of Total Bill Amounts')
plt.xlabel('Total Bill ($)')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

**Box Plot**
- Plot the distribution of total bill amounts by day.

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='day', y='total_bill', data=tips)
plt.title('Total Bill Amounts by Day')
plt.xlabel('Day of the Week')
plt.ylabel('Total Bill ($)')
plt.show()
```

**Subplots**
- Create a 2x2 grid of subplots for different plots.

```python
fig, axs = plt.subplots(2, 2, figsize=(12, 12))

# Scatter plot
axs[0, 0].scatter(tips['total_bill'], tips['tip'], c='blue')
axs[0, 0].set_title('Total Bill vs Tip')
axs[0, 0].set_xlabel('Total Bill ($)')
axs[0, 0].set_ylabel('Tip ($)')

# Histogram
axs[0, 1].hist(tips['total_bill'], bins=30, color='purple', edgecolor='black')
axs[0, 1].set_title('Distribution of Total Bill Amounts')
axs[0, 1].set_xlabel('Total Bill ($)')
axs[0, 1].set_ylabel('Frequency')

# Box plot
sns.boxplot(x='day', y='total_bill', data=tips, ax=axs[1, 0])
axs[1, 0].set_title('Total Bill Amounts by Day')
axs[1, 0].set_xlabel('Day of the Week')
axs[1, 0].set_ylabel('Total Bill ($)')

# Bar plot
day_counts = tips['day'].value_counts()
axs[1, 1].bar(day_counts.index, day_counts.values, color='orange')
axs[1, 1].set_title('Count of Days')
axs[1, 1].set_xlabel('Day of the Week')
axs[1, 1].set_ylabel('Count')

plt.tight_layout()
plt.show()
```

**Dataset**
- Wine Quality Dataset - Available from the UCI Machine Learning Repository.

**Example 1: Histogram**
- Plot the distribution of wine quality scores.

```python
import pandas as pd

wine = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/winequality-red.csv', delimiter=';')

plt.figure(figsize=(10, 6))
plt.hist(wine['quality'], bins=10, color='purple', edgecolor='black')
plt.title('Distribution of Wine Quality Scores')
plt.xlabel('Quality Score')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

**Example 2: Box Plot**
- Plot the distribution of alcohol content by wine quality.

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='quality', y='alcohol', data=wine)
plt.title('Distribution of Alcohol Content by Wine Quality')
plt.xlabel('Wine Quality')
plt.ylabel('Alcohol Content (%)')
plt.show()
```

**Example 3: Pair Plot**
- Create a pair plot to visualize relationships between alcohol, pH, and quality.

```python
selected_features = wine[['alcohol', 'pH', 'quality']]
sns.pairplot(selected_features, hue='quality', palette='coolwarm')
plt.suptitle('Pair Plot of Selected Wine Features', y=1.02)
plt.show()
```

**Dataset**
- Boston Housing Dataset - Available from the UCI Machine Learning Repository or the sklearn library.

**Example 1: Heatmap**
- Create a heatmap of the correlation matrix of the Boston Housing dataset.

```python
from sklearn.datasets import fetch_california_housing
import seaborn as sns
import pandas as pd

# Load the California housing dataset
california = fetch_california_housing()

# Convert to a DataFrame for easier handling
df_california = pd.DataFrame(california.data, columns=california.feature_names)
df_california['target'] = california.target

# Display the first few rows
df_california.head()
```

```python
# Example visualization with seaborn
sns.pairplot(df_california, diag_kind='kde')
```

```python
from sklearn.datasets import fetch_california_housing
import seaborn as sns

california = fetch_california_housing()
california_df = pd.DataFrame(california.data, columns=california.feature_names)
california_df['PRICE'] = california.target

plt.figure(figsize=(12, 10))
corr_matrix = california_df.corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix of California Housing Dataset')
plt.show()
```
