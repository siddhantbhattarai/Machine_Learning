**Beginner Level Solutions**
- Question 1: Line Plot

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
titanic = pd.read_csv('https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv')

# Cumulative sum of passenger ages
titanic['Cumulative Age'] = titanic['Age'].cumsum()

plt.figure(figsize=(10, 6))
plt.plot(titanic.index, titanic['Cumulative Age'], marker='o')
plt.title('Cumulative Sum of Passenger Ages')
plt.xlabel('Passenger Index')
plt.ylabel('Cumulative Age')
plt.grid(True)
plt.show()
```

Question 2: Scatter Plot

```python
plt.figure(figsize=(10, 6))
plt.scatter(titanic['Age'], titanic['Fare'], alpha=0.5)
plt.title('Age vs. Fare')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.grid(True)
plt.show()
```

Question 3: Bar Chart

```python
survivors = titanic['Survived'].value_counts()

plt.figure(figsize=(8, 6))
plt.bar(['Not Survived', 'Survived'], survivors, color=['red', 'green'])
plt.title('Survivors Count')
plt.xlabel('Survival Status')
plt.ylabel('Count')
plt.show()
```

Question 6: Simple Line Plot

```python
temperature = pd.read_csv('https://raw.githubusercontent.com/jbrownlee/Datasets/master/daily-min-temperatures.csv')

plt.figure(figsize=(10, 6))
plt.plot(temperature['Date'], temperature['Temp'], color='blue')
plt.title('Daily Temperatures')
plt.xlabel('Date')
plt.ylabel('Temperature')
plt.grid(True)
plt.show()
```
