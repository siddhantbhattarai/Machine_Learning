**Introduction and Setup**

**Dataset**
- Stock Market Dataset - Use the Yahoo Finance API or a CSV file of historical stock prices (e.g., Tesla stock prices).


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load data
df = pd.read_csv('https://raw.githubusercontent.com/siddhantbhattarai/Machine_Learning/main/Datasets/TSLA.csv')

# Convert 'Date' column to datetime
df['Date'] = pd.to_datetime(df['Date'])

df.head()
```

**Custom Line Plot with Annotations**
- Plot the closing prices with annotations for key events.


```python
# Create the plot
plt.figure(figsize=(14, 7))
plt.plot(df['Date'], df['Close'], label='Closing Price')
plt.title('Tesla Stock Prices')
plt.xlabel('Date')
plt.ylabel('Closing Price ($)')
plt.grid(True)
plt.legend()

# Annotate with correct datetime format
plt.annotate('COVID-19 Pandemic',
             xy=(pd.to_datetime('2020-03-02'), 160),
             xytext=(pd.to_datetime('2020-01-02'), 400),
             arrowprops=dict(facecolor='black', shrink=0.05))

plt.show()
```

**3D Plot**
- Plot a 3D scatter plot of open, high, and close prices.

```python
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(df['Open'], df['High'], df['Close'], c='red', marker='o')
ax.set_title('3D Scatter Plot of Stock Prices')
ax.set_xlabel('Open Price ($)')
ax.set_ylabel('High Price ($)')
ax.set_zlabel('Close Price ($)')
plt.show()
```

**Custom Animation**
- Create an animated plot of closing prices over time.

```python
from matplotlib.animation import FuncAnimation

# Create the figure and axis
fig, ax = plt.subplots(figsize=(14, 7))

# Set x and y limits based on data
ax.set_xlim(df['Date'].min(), df['Date'].max())
ax.set_ylim(df['Close'].min(), df['Close'].max())

# Initialize line plot
line, = ax.plot([], [], color='blue')

def init():
    line.set_data([], [])
    return line,

def animate(i):
    line.set_data(df['Date'][:i], df['Close'][:i])
    return line,

# Create animation
anim = FuncAnimation(fig, animate, init_func=init, frames=len(df), interval=50, blit=False)

# Display the plot
plt.title('Animated Plot of Tesla Closing Prices')
plt.xlabel('Date')
plt.ylabel('Closing Price ($)')

# Save animation to a file (optional)
anim.save('tesla_animation.gif', writer='pillow')

plt.show()
```
